```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

import sys
sys.path.append("..")
```

```python
survey_df = pd.read_csv('../data/laurelhurst_survey_with_address.csv')
survey_df
```

```python
# Select the columns
columns = ['7_water', '7_food', '7_meds', '7_transp', '7_comm', '7_faid', '7_warm', '7_sani', '7_power', '7_shelter']
corr_matrix = survey_df[columns].corr()
corr_matrix
```

```python
# get the PMF of each column of the data
sharing_pref_pmf = survey_df[columns].apply(pd.value_counts, normalize=True)
sharing_pref_pmf
```

```python
# find the covariance matrix of the generated RVs
cov_mat = survey_df[columns].cov()
cov_mat
```

```python
# Cholesky decomposition
L = np.linalg.cholesky(cov_mat)
L
```

```python
# Check Cholesky decomposition
L @ L.T
```

```python
# Generate uncorrelated random variables
uncorr_sharing_preferences = np.zeros((10, 1000))
np.random.seed(0)
for i,rsc in enumerate(columns):
    for j in range(uncorr_sharing_preferences.shape[1]):
        uncorr_sharing_preferences[i, j] = np.random.choice(sharing_pref_pmf.index, p=sharing_pref_pmf[rsc].values)
uncorr_sharing_preferences
```

```python
print(np.corrcoef(uncorr_sharing_preferences).round(2))
```

```python
print(np.cov(uncorr_sharing_preferences).round(2))
```

```python
# Generate correlated random variables
corr_sharing_pref = L @ uncorr_sharing_preferences
corr_sharing_pref
```

```python
print(np.corrcoef(corr_sharing_pref).round(2))
```
